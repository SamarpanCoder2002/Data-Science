```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv(r"""F:\Learning Files\data_science\dataset\Dataset\data1\Travel.csv""");
```

```python
df
```

```python
# Finding missing values
missing_values = [col for col in df.columns if df[col].isnull().sum() > 0]
```

```python
missing_values
```

```python
# Length of missing value
len(missing_values)
```

```python
# Check null values in graph
plt.figure(figsize=(20,10))
sns.heatmap(df.isnull())
```

```python
# Droping a particular column
df.drop('CustomerID', axis=1) # Dropping a column
```

```python
# Numerical columns filtration
num_data_cols = [col for col in df.columns if df[col].dtypes != 'O']
```

```python
num_data_cols
```

```python
# Categorial Columns Filtration
cat_data_cols = [col for col in df.columns if df[col].dtypes == 'O']
```

```python
cat_data_cols
```

```python
# Getting numerical data
num_data = df[num_data_cols]
```

```python
num_data
```

```python
# Heat map to show in graph about numerical value with Empty data
plt.figure(figsize=(20,10))
sns.heatmap(num_data.isnull())
```

```python
# Basic statistical operation on numerical data
num_data.describe().T
```

```python
# Checking correlation on numerical data
num_data.corr() # Correlation on numerical data
```

```python
# Checking Covariance on numerical data
num_data.cov() # Covariance on numerical data
```

```python
# Distribution ploting of particular numerical column
plt.figure(figsize=(20, 10))
sns.displot(num_data['Age']) # From that distribution plot we can understand mean
```
